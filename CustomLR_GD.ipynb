```python
class MyLR_GD:
  def __init__(self, lr, epochs):
    self.m = 972
    self.b = 150.26
    self.lr = lr
    self.epochs = epochs

  def fit(self, X_train, y_train):
    n = len(X_train)
    for epoch in range(self.epochs):
      y_pred = self.m * X_train + self.b
      dm = (2/n) * np.sum(X_train * (y_pred - y_train))
      db = (2/n) * np.sum(y_pred - y_train)
      self.m = self.m - self.lr * dm
      self.b = self.b - self.lr * db
    print(self.m, self.b)

  def predict(self, X_test):
    return self.m * X_test + self.b
```

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes
```

```python
X, y = load_diabetes(return_X_y=True)
```

```python
X_single_feature = X[:, 2].reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_single_feature, y, test_size=0.33, random_state=2)
```

```python
gd_lr = MyLR_GD(0.001, 1000)
```

```python
gd_lr.fit(X_train, y_train)
```

```python
y_pred = gd_lr.predict(X_test)
```
